```python
import tensorflow as tf
```

```python
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
from sklearn.linear_model import *
```

```python
train_data=pd.read_csv("train_data.csv")
data=train_data
```

```python
data
```

```python
trainx=[[23.90,30.10,74.20,94.60,0.6,2.6,24,29,76,85,0,6,0.8]]
trainy=[3.5]
for i in range(1,344):
    trainx.append([data['tmin'][i],data['tmax'][i],data['hmin'][i],data['hmax'][i],data['wmin'][i],data['wmax'][i],
                   data['tmin'][i-1],data['tmax'][i-1],data['hmin'][i-1],data['hmax'][i-1],data['wmin'][i-1],data['wmax'][i-1],data['r'][i-1]])
    trainy.append(data['r'][i])
```

```python
xt=[]
for i in range(0,70):
    xt.append(i)
```

```python
m=tf.keras.models.Sequential([tf.kears.layers.
    tf.keras.layers.Dense(1,activation='relu'),
                   tf.keras.layers.Dense(1024,activation='relu'),
                   tf.keras.layers.Dense(512,activation='relu'),
                   tf.keras.layers.Dense(1024,activation='relu'),
                   tf.keras.layers.Dense(1024,activation='relu'),
                   tf.keras.layers.Dense(1,activation='linear')])
m.compile(optimizer='adam',loss='mean_squared_error',metrics='mse')
m.fit(trainx,trainy,epochs=1000)
```
